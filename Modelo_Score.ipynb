### Importación de paqueterías

### Descarga de base de datos

```python
data = pd.read_csv('Base Premier League.csv')
data
```

### Limpieza de Datos

```python
cd.dqr(data)
```

```python
data["PRESUPUESTO PARA TRASNFERENCIAS"] = data["PRESUPUESTO PARA TRASNFERENCIAS"].apply(cd.remove_punctuation).astype(float)
data["PRESUPUESTO PARA SUELDOS"] = data["PRESUPUESTO PARA SUELDOS"].apply(cd.remove_punctuation).astype(float)
data["INGRESOS POR TV"] = data["INGRESOS POR TV"].apply(cd.remove_punctuation).astype(float)
data["BONO POR POSICION"] = data["BONO POR POSICION"].apply(cd.remove_punctuation).astype(float)
data["BONOS CHAMPIONS LEAGUE"] = data["BONOS CHAMPIONS LEAGUE"].apply(cd.remove_punctuation).astype(float)
data["BONOS EUROPA LEAGUE"] = data["BONOS EUROPA LEAGUE"].apply(cd.remove_punctuation).astype(float)
data["BONOS CONFERENCE"] = data["BONOS CONFERENCE"].apply(cd.remove_punctuation).astype(float)
data["INGRESO POR ASCENSO"] = data["INGRESO POR ASCENSO"].apply(cd.remove_punctuation).astype(float)
data["INGRESOS POR TRASPASOS"] = data["INGRESOS POR TRASPASOS"].apply(cd.remove_punctuation).astype(float)
data["GASTOS POR TRASPASOS"] = -1*(data["GASTOS POR TRASPASOS"].apply(cd.remove_punctuation).astype(float))
data["ASISTENCIA PROMEDIO"] = data["ASISTENCIA PROMEDIO"].apply(cd.remove_punctuation).astype(float)
data["PRECIO DE ENTRADA"] = data["PRECIO DE ENTRADA"].apply(cd.remove_punctuation).astype(float)
data["GANANCIAS ESTADIO"] = data["GANANCIAS ESTADIO"].apply(cd.remove_punctuation).astype(float)
data = data.drop(["Unnamed: 16","Unnamed: 17"], axis=1)
data
```

```python
cd.dqr(data)
```

### Analisis Exploratorio de los Datos

#### Matriz de Correlación

```python
# Convertir la columna SCORE_MS a una variable categórica ordinal
score_mapping = {'EXCELENTE': 3, 'BUENO': 2, 'MALO': 1}
data['SCORE_NUMERIC'] = data['SCORE_MS'].map(score_mapping)

# Calcular correlaciones con la calificación numérica
correlation_matrix = data.corr()

# Visualizar la matriz de correlación
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()

# Mostrar correlaciones específicas con SCORE_NUMERIC
correlation_with_score = correlation_matrix['SCORE_NUMERIC'].sort_values(ascending=False)
correlation_with_score
```

```python
correlation_with_score
```

#### Asignar pesos a las variables

```python
# Asignar los pesos a cada variable
weights = {
    'BONO POR POSICION': 0.20,
    'BONOS CHAMPIONS LEAGUE': 0.15,
    'PRESUPUESTO PARA SUELDOS': 0.10,
    'GASTOS POR TRASPASOS': 0.11,
    'PRESUPUESTO PARA TRASNFERENCIAS': 0.01,
    'BONOS EUROPA LEAGUE': 0.10,
    'INGRESOS POR TRASPASOS': 0.07,
    'BONOS CONFERENCE': 0.10,
    'INGRESO POR ASCENSO': 0.05,
    'GANANCIAS ESTADIO': 0.10,
    'INGRESOS POR TV': 0.01
    
}

# Normalizar las variables para que tengan el mismo rango antes de aplicar los pesos
normalized_data = data[weights.keys()].apply(lambda x: (x - x.min()) / (x.max() - x.min()))

# Calcular el índice de crédito ponderado
data['CREDIT_INDEX'] = normalized_data.mul(weights).sum(axis=1)

# Verificar el índice calculado
data[['EQUIPO', 'SCORE_MS', 'CREDIT_INDEX']]
```

#### Distribución de los datos

```python
# Explorar la distribución del índice de crédito
plt.figure(figsize=(10, 6))
sns.histplot(data['CREDIT_INDEX'], kde=True, bins=20)
plt.title("Distribución del Índice de Crédito")
plt.xlabel("Índice de Crédito")
plt.ylabel("Frecuencia")
plt.show()

# Calcular percentiles para sugerir umbrales
percentiles = data['CREDIT_INDEX'].quantile([0.33, 0.66]).values
percentiles
```

### Modelo básico de Crédito

```python
# Aplicar los umbrales para clasificar a cada equipo
def classify_credit_score(index):
    if index <= .25:
        return 'MALO'
    elif index <= .39:
        return 'BUENO'
    else:
        return 'EXCELENTE'

data['MODELED_SCORE'] = data['CREDIT_INDEX'].apply(classify_credit_score)

# Comparar las nuevas calificaciones con las originales
comparison = data[['EQUIPO', 'SCORE_MS', 'MODELED_SCORE', 'CREDIT_INDEX']]
comparison
```

#### Coincidencia de los resultados

```python
coincidencias = (data['SCORE_MS'] == data['MODELED_SCORE']).sum()
porcentaje_coincidencia = (coincidencias / len(data))*100

print(f"El porcentaje de coincidencia entre las dos columnas es: {porcentaje_coincidencia:.2f}%")
```
